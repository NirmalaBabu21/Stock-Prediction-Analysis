```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas_datareader.data as web
from datetime import datetime
```

```python
end_date = datetime.now()
start_date = datetime(end_date.year-1,end_date.month,end_date.day)
print("Start Date:",start_date)
print("End Date: ", end_date)
```

```python
#getting the datas from yahoo source
apple_data = web.DataReader(name='AAPL',data_source='yahoo',start=start_date,end=end_date)
microsoft_data = web.DataReader(name='MSFT',data_source='yahoo',start=start_date,end=end_date)
google_data = web.DataReader(name='GOOG',data_source='yahoo',start=start_date,end=end_date)
amazon_data = web.DataReader(name='AMZN',data_source='yahoo',start=start_date,end=end_date)

apple_data.columns = 'Apple_' + apple_data.columns
microsoft_data.columns = 'Microsoft_' +microsoft_data.columns
google_data.columns = 'Google_'+ google_data.columns
amazon_data.columns = 'Amazon_'+ amazon_data.columns

apple_data['company'] = 'Apple'
microsoft_data['company'] = 'Microsoft'
google_data['company'] = 'Google'
amazon_data['company'] = 'Amazon'
```

```python
apple_data['Apple_MA for 20 Days'] = apple_data['Apple_Adj Close'].rolling(20).mean()
microsoft_data['Microsoft_MA for 20 Days'] = microsoft_data['Microsoft_Adj Close'].rolling(20).mean()
google_data['Google_MA for 20 Days'] = google_data['Google_Adj Close'].rolling(20).mean()
amazon_data['Amazon_MA for 20 Days'] = amazon_data['Amazon_Adj Close'].rolling(20).mean()


apple_data['Apple_MA for 50 Days'] = apple_data['Apple_Adj Close'].rolling(50).mean()
microsoft_data['Microsoft_MA for 50 Days'] = microsoft_data['Microsoft_Adj Close'].rolling(50).mean()
google_data['Google_MA for 50 Days'] = google_data['Google_Adj Close'].rolling(50).mean()
amazon_data['Amazon_MA for 50 Days'] = amazon_data['Amazon_Adj Close'].rolling(50).mean()

apple_data['Apple_MA for 100 Days'] = apple_data['Apple_Adj Close'].rolling(100).mean()
microsoft_data['Microsoft_MA for 100 Days'] = microsoft_data['Microsoft_Adj Close'].rolling(100).mean()
google_data['Google_MA for 100 Days'] = google_data['Google_Adj Close'].rolling(100).mean()
amazon_data['Amazon_MA for 100 Days'] = amazon_data['Amazon_Adj Close'].rolling(100).mean()

apple_data['Apple_Daily Returns'] = apple_data['Apple_Adj Close'].pct_change()
microsoft_data['Microsoft_Daily Returns'] = microsoft_data['Microsoft_Adj Close'].pct_change()
google_data['Google_Daily Returns'] = google_data['Google_Adj Close'].pct_change()
amazon_data['Amazon_Daily Returns'] = amazon_data['Amazon_Adj Close'].pct_change()
```

```python
apple_data.head()
```

```python
def trend(x):
  if x > -0.015 and x <= 0.015:
    return 'Slight or No change'
  elif x > 0.015 and x <= 0.04:
    return 'Slight Positive'
  elif x < -0.015 and x >= -0.4:
    return 'Slight Negative'
  elif x > 0.04 and x <= 0.06:
    return 'Positive'
  elif x < -0.04 and x >= -0.06:
    return 'Negative'
  elif x > 0.06 and x <= 0.07:
    return 'Among top gainers'
  elif x < -0.06 and x >= -0.07:
    return 'Among top losers'
  elif x > 0.07:
    return 'Bull run'
  elif x <= -0.07:
    return 'Bear drop'

apple_data['apple_Trend'] = apple_data['Apple_Daily Returns'].apply(lambda x: trend(x))
microsoft_data['microsoft_Trend'] = microsoft_data['Microsoft_Daily Returns'].apply(lambda x: trend(x))
google_data['google_Trend'] = google_data['Google_Daily Returns'].apply(lambda x: trend(x))
amazon_data['amazon_Trend'] = amazon_data['Amazon_Daily Returns'].apply(lambda x: trend(x))
```

```python
df = pd.concat([apple_data, microsoft_data, google_data, amazon_data], axis=1).reset_index()
```
